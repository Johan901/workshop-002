```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

spotify_df = pd.read_csv('spotify_dataset.csv')
grammy_df = pd.read_csv('the_grammy_awards.csv')

# EDA para Spotify Dataset
print("EDA para el Dataset de Spotify")
print("="*40)

print("Primeras 5 filas del dataset de Spotify:")
print(spotify_df.head())

print("\nDimensiones del dataset de Spotify:", spotify_df.shape)
print("\nTipos de datos en el dataset de Spotify:")
print(spotify_df.dtypes)

print("\nResumen estadístico de las columnas numéricas en Spotify:")
print(spotify_df.describe())

print("\nNúmero de valores nulos por columna en Spotify:")
print(spotify_df.isnull().sum())

plt.figure(figsize=(10,6))
sns.histplot(spotify_df['popularity'], bins=30, kde=True, color='blue')
plt.title('Distribución de la Popularidad de las Canciones (Spotify)')
plt.xlabel('Popularidad')
plt.ylabel('Frecuencia')
plt.show()

# EDA para Grammy Awards Dataset
print("\nEDA para el Dataset de los Grammy Awards")
print("="*40)

print("Primeras 5 filas del dataset de Grammy Awards:")
print(grammy_df.head())

print("\nDimensiones del dataset de Grammy Awards:", grammy_df.shape)
print("\nTipos de datos en el dataset de Grammy Awards:")
print(grammy_df.dtypes)

print("\nResumen estadístico de las columnas numéricas en Grammy Awards:")
print(grammy_df.describe())

print("\nNúmero de valores nulos por columna en Grammy Awards:")
print(grammy_df.isnull().sum())

plt.figure(figsize=(14,6))
sns.countplot(data=grammy_df, x='year', hue='winner', palette='Set2')
plt.title('Ganadores de los Grammy por Año')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

print("\nTop 10 categorías en los Grammy Awards:")
print(grammy_df['category'].value_counts().head(10))

plt.figure(figsize=(12,6))
grammy_df['category'].value_counts().head(10).plot(kind='bar', color='green')
plt.title('Top 10 Categorías más Premiados (Grammy Awards)')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

print("\nTop 10 artistas con más nominaciones en los Grammy Awards:")
print(grammy_df['artist'].value_counts().head(10))

plt.figure(figsize=(12,6))
grammy_df['artist'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Artistas con Más Nominaciones (Grammy Awards)')
plt.xlabel('Artista')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

cross_tab = pd.crosstab(grammy_df['artist'], grammy_df['category'])
print("\nAnálisis cruzado de artistas y categorías:")
print(cross_tab.head())

# Finalización del EDA
print("\nEDA completado.")
```
